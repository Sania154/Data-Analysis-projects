```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,440,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,histtype='bar',orientation='vertical',color='r',rwidth=0.8,
        label='Ml student')
plt.title('Ml student age histrogram')
plt.xlabel('ml_student_age')
plt.ylabel('no student age')
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,histtype='bar',orientation='horizontal',color='k',rwidth=0.8,
        label='Ml student')
plt.title('Ml student age histrogram')
plt.xlabel('ml_student_age,histtype')
plt.ylabel('no student age')
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,10))
plt.hist([ml_student_age,py_student_age],bins,histtype='bar',orientation='vertical',color=['y','m'],rwidth=0.8,
        label=['ml student','py student'])
plt.title('ML student age histrogram')
plt.xlabel('ml_student_age')
plt.ylabel('no student age')
plt.legend()
plt.grid(color='k',linestyle='-',linewidth=1)
plt.show()
```
