# Skiprows

```python
df=pd.read_csv('C:\\Users\\Super\\Downloads\\sania_1xls.csv.2.csv',skiprows=2)
df
```

```python
df=pd.read_csv('C:\\Users\\Super\\Downloads\\sania_1xls.csv.2.csv',nrows=10,usecols=[0,1,2,3],skiprows=[0])
df
```

```python
df=pd.read_csv('C:\\Users\\Super\\Downloads\\sania_1xls.csv.2.csv',nrows=10,usecols=[0,1,2,3],skiprows=[1])
df
```

```python
df=pd.read_csv('C:\\Users\\Super\\Downloads\\sania_1xls.csv.2.csv',nrows=10,usecols=[0,1,2,3],skiprows=[1,3,5])
df
```
