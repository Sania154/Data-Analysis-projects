```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=['python','R','AI','ML','DS']
class1=[30,10,20,25,10]
class2=[40,5,20,20,10]
class3=[35,5,30,15,15]
```

```python
plt.bar(classes,class1)
plt.show()
```

```python
plt.barh(classes,class1)
plt.show()
```

```python
plt.figure(figsize=(10,10))
style.use('ggplot')
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1,width,color='b', label='class1 student')

plt.bar(class_index+width,class2,width,color='g', label='class2 student')

plt.bar(class_index+width+width,class3,width,color='y', label='class3 student')

plt.title('Bar chart os IAIP C lass',fontsize=16)
plt.xlabel('classes',fontsize=14)
plt.ylabel('No of student',fontsize=14)
plt.legend(loc=1)
plt.xticks(class_index+width,classes,rotation=30)
plt.show()
```
